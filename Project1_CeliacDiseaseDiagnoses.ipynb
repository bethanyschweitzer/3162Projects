```python
import pandas as pd
celiac_df=pd.read_csv("/Users/admin/Downloads/celiac_disease_lab_data.csv")
import seaborn as sns
sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt
```

```python
# Drawing a nested barplot by diabetes and celiac disease diagnoses
g = sns.catplot(
    data=celiac_df, kind="count",
    x="Disease_Diagnose", hue="Diabetes",
    palette="dark", alpha=.6,height = 6
)
```

```python
sns.histplot(data=celiac_df, x="Disease_Diagnose", hue="Diabetes", stat='probability', multiple="dodge", shrink=.8)
```

```python
g = sns.catplot(
    data=celiac_df, kind="count",
    x="Disease_Diagnose", hue="Abdominal",
    palette="dark", alpha=.6,height = 6
)
```

```python
# Calculate percentages
percentage_yes_yes = (celiac_df[(celiac_df['Disease_Diagnose'] == 'yes') & (celiac_df['Abdominal'] == 'yes')].shape[0] /
                      celiac_df[celiac_df['Disease_Diagnose'] == 'yes'].shape[0]) * 100

percentage_no_yes = (celiac_df[(celiac_df['Disease_Diagnose'] == 'no') & (celiac_df['Abdominal'] == 'yes')].shape[0] /
                     celiac_df[celiac_df['Disease_Diagnose'] == 'no'].shape[0]) * 100

# Create a bar plot
sns.barplot(x=['Yes', 'No'], y=[percentage_yes_yes, percentage_no_yes], palette="dark", alpha=0.6)

# Set labels and title
plt.xlabel("Disease Diagnosis")
plt.ylabel("Percentage")

# Show the plot
plt.show() 
```
